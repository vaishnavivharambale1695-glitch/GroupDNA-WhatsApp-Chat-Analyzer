```python

import numpy as np
from datetime import datetime
# File Path

file_path = r"C:\Users\vaish\Downloads\12721105-DADS_Minor_PROJECT_dataset\DADS Minor PROJECT dataset.txt"
with open(file_path, "r", encoding="utf-8") as file:
    lines = file.readlines()

print("Total Lines :", len(lines))
messages = []

participants = set()

system_count = 0

media_count = 0

deleted_count = 0
previous_message = None

for line in lines:

    line = line.strip()

    if line == "":
        continue

    
    if len(line) < 8 or line[2] != "/" or line[5] != "/":

        if previous_message is not None:

            previous_message["text"] += " " + line

        continue

    if " - " not in line:

        continue

    timestamp, remaining = line.split(" - ", 1)

    if ": " not in remaining:

        system_count += 1
        continue

    sender, text = remaining.split(": ", 1)

    participants.add(sender)

    if text == "<Media omitted>":

        media_count += 1

    elif text == "This message was deleted":

        deleted_count += 1

    message = {

        "timestamp": timestamp,
        "sender": sender,
        "text": text

    }

    messages.append(message)

    previous_message = message


print("Parsing Completed Successfully")



print("Total Parsed Messages :", len(messages))

print("Participants :", len(participants))

print("System Messages :", system_count)

print("Media Messages :", media_count)

print("Deleted Messages :", deleted_count)


print("\nFirst Five Messages\n")

for msg in messages[:5]:

    print(msg)
    print("\nLast Five Messages\n")

for msg in messages[-5:]:

    print(msg)
```

```python
total_messages = len(messages)

print("Total Messages :", total_messages)

total_participants = len(participants)

print("Participants :", total_participants)

first_date = datetime.strptime(messages[0]["timestamp"], "%d/%m/%y, %H:%M")
last_date = datetime.strptime(messages[-1]["timestamp"], "%d/%m/%y, %H:%M")

print("Start Date :", first_date.strftime("%d %B %Y"))
print("End Date   :", last_date.strftime("%d %B %Y"))

total_days = (last_date - first_date).days + 1

print("Total Days :", total_days)

message_count = {}

for msg in messages:

    sender = msg["sender"]

    if sender in message_count:

        message_count[sender] += 1

    else:

        message_count[sender] = 1

sorted_people = sorted(
    message_count.items(),
    key=lambda x: x[1],
    reverse=True
)
print("=" * 60)
print("GROUP OVERVIEW")
print("=" * 60)

print(f"Group Name      : Hostel Bois 4ever")
print(f"Period          : {first_date.strftime('%d %B %Y')} to {last_date.strftime('%d %B %Y')}")
print(f"Duration        : {total_days} Days")
print(f"Total Messages  : {total_messages}")
print(f"Participants    : {total_participants}")

print("=" * 60)
print("MESSAGES PER PERSON")
print("=" * 60)

for person, count in sorted_people:

    percentage = (count / total_messages) * 100

    print(f"{person:<10} {count:>5} ({percentage:.1f}%)")
    

```

```python


day_count = {}

for msg in messages:

    date = msg["timestamp"].split(",")[0]

    if date in day_count:
        day_count[date] += 1
    else:
        day_count[date] = 1
 
busiest_day = max(day_count, key=day_count.get)

busiest_day_messages = day_count[busiest_day]

formatted_day = datetime.strptime(
    busiest_day,
    "%d/%m/%y"
).strftime("%d %B %Y")

hour_count = {}

for hour in range(24):
    hour_count[hour] = 0

for msg in messages:

    time = msg["timestamp"].split(", ")[1]

    hour = int(time.split(":")[0])

    hour_count[hour] += 1

busiest_hour = max(hour_count, key=hour_count.get)

busiest_hour_messages = hour_count[busiest_hour]

next_hour = (busiest_hour + 1) % 24
print("=" * 60)
print("MOST ACTIVE TIME")
print("=" * 60)

print(
    f"Busiest Day  : {formatted_day} ({busiest_day_messages} messages)"
)

print(
    f"Busiest Hour : {busiest_hour:02d}:00 - {next_hour:02d}:00 ({busiest_hour_messages} messages)"
)

print("=" * 60)
print("\nMESSAGES BY HOUR")
print("=" * 60)

for hour in range(24):

    next_hour = (hour + 1) % 24

    print(
        f"{hour:02d}:00 - {next_hour:02d}:00 : {hour_count[hour]}"
    )
    


    
```

```python


people = sorted(list(participants))

print("Participants:")

for person in people:
    print(person)

person_index = {}

for i in range(len(people)):
    person_index[people[i]] = i

print(person_index)

heatmap = np.zeros((len(people), 24), dtype=int)

print(heatmap)

for msg in messages:

    sender = msg["sender"]

    row = person_index[sender]

    time = msg["timestamp"].split(", ")[1]

    hour = int(time.split(":")[0])

    heatmap[row][hour] += 1
print("Heatmap Matrix")

print(heatmap)
print("=" * 90)
print("ACTIVITY HEATMAP")
print("=" * 90)

print("Name       ", end="")

for hour in range(24):
    print(f"{hour:02d} ", end="")

print()
for i in range(len(people)):

    print(f"{people[i]:<10}", end=" ")

    maximum = heatmap[i].max()

    if maximum == 0:
        maximum = 1

    for value in heatmap[i]:

        ratio = value / maximum

        if ratio == 0:
            symbol = ". "

        elif ratio <= 0.25:
            symbol = "░ "

        elif ratio <= 0.50:
            symbol = "▒ "

        else:
            symbol = "█ "

        print(symbol, end="")

    print()
    print("\n")

print("=" * 60)

print("TOTAL MESSAGES FROM MATRIX")

print("=" * 60)

for i in range(len(people)):

    total = np.sum(heatmap[i])

    print(f"{people[i]:<10} {total}")

    

```

```python


stop_words = [

    "i","is","the","a","an","and","or","to","of","in",
    "on","for","at","it","its","it's","are","am","was",
    "were","be","been","being","you","me","my","we",
    "our","your","this","that","with","from","as","if",
    "but","so","do","does","did"

]

punctuation = ".,!?;:()[]{}<>\"'`-_/\\|@#$%^&*+=~"

word_count = {}

for msg in messages:

    text = msg["text"]

    # Ignore media
    if text == "<Media omitted>":
        continue

    # Ignore deleted messages
    if text == "This message was deleted":
        continue

    words = text.split()

    for word in words:

        word = word.lower()

        word = word.strip(punctuation)

        if word == "":
            continue

        if word in stop_words:
            continue

        if word in word_count:

            word_count[word] += 1

        else:

            word_count[word] = 1

top_words = sorted(

    word_count.items(),

    key=lambda x: x[1],

    reverse=True

)

print("=" * 70)

print("THIS GROUP'S FAVOURITE WORDS")

print("=" * 70)

maximum = top_words[0][1]

for word, count in top_words[:10]:

    bar = "█" * int((count / maximum) * 25)

    print(f"{word:<15} {bar:<25} {count}")    
    
```
